## Task 1: Pandas series basics

````python
marks = pd.Series([78, 80, 90, 66, 72])

## Printing the series
print(marks)
````

````python
## Printing the indexes of the series
print(marks.index)
````

````python
## Printing the data type of the series
print(marks.dtype)
````

````python
## Accessing elements using index
print("Element at index 2:", marks[2])
print("Element at index 3,4:", marks[3:5])
````

## Mathematical operations on series

````python
## Adding 5 grace marks to each student's score
print(marks.add(5))
## Subtracting 2 marks from each student's score
print(marks.subtract(2))
## Multiplying each student's score by 1.05 (to give a 5% bonus)
print(marks.multiply(1.05))
## Dividing each student's score by 2
print(marks.divide(2))
````

## Task 3: Python functionalities with pandas series

````python
## Finding the max of all marks
print(marks.max())
## Finding the min of all marks
print(marks.min())
## Findig the sum of all marks
print(marks.sum())
## Finding the mean of all marks
print(marks.mean())

## Lambda function to check who scored more than or equal to 70
print("Total number of students passed:", marks.apply(lambda x: x >= 70).sum())
````

## Task 4 Creating a dataframe from series

````python

students = {
    'Name': ["Amit", "Neha", "Rahul", "Sneha", "Pooja"],
    'Marks': [78, 85, 90, 66, 72],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math']
}

df = pd.DataFrame(students)
print(df)
````

````python
df.iloc[1:4]
````

````python
df.iloc[-2:]
````

````python
print(df.shape)
print(df.columns)
````

## Task 5: Important Dataframe Functions

````python
# Printing dataframe info
print(df.info())

## Printing dataframe statistical summary
print(df.describe())

## Printing dataframe head (first 3 rows)
print(df.head())

## Printing dataframe tail (last 2 rows)
print(df.tail())
````

## Task 6: Filtering and Conditional Selection

````python
print("Students who scored more than 75% marks:")
print(df[df['Marks'] > 75])
print("\n")
print("Students belonging to subject math:")
print(df[df['Subject'] == "Math"])
print("\n")
print("Students who scored more than average marks")
print(df[df['Marks'] > df['Marks'].mean()])
print("\n")
print("Students who failed (marks<70)")
print(df[df['Marks'] < 70])
print("\n")
````

## Task 7: Grouping and Basic Analysis

````python
print("Average marks obtained in each subject:")
print(df.groupby('Subject')['Marks'].mean())
print("\n")
print("Number of students in each subject:")
print(df['Subject'].value_counts())
print("\n")
print("Maximum marks obtained in each subject:")
print(df.groupby('Subject')['Marks'].max())
````

## Task 9: Sales data analysis

````python
sales = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Revenue': [1200, 1500, 900, 2000, 1800]
}

df = pd.DataFrame(sales)
df.head()
````

````python
df.info()
````

````python
print("The total sales revenue:", df['Revenue'].sum())
print("Average sales revenue:", df['Revenue'].mean())
print("Days where revenue greater than average revenue\n", df[df['Revenue']>df['Revenue'].mean()])
df.plot(x='Day', y='Revenue', kind = 'bar')
````
